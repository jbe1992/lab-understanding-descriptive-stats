# Challenge 1

```python
from random import choice
import pandas as pd
%matplotlib inline
```

```python
dice = [1, 2, 3, 4, 5, 6]
```

```python
dice_results = pd.DataFrame([choice(dice) for x in range(10)])
```

```python
dice_results.rename(columns = {0 : "Roll"}, inplace = True)
```

```python
dice_results.head(10)
```

```python
dice_results.sort_values('Roll').plot(kind ='bar')
```

```python
freq_df = pd.DataFrame(dice_results['Roll'].value_counts())
```

```python
freq_df = freq_df.reset_index().rename(columns={'Roll':'Frequency','index':'Roll'})
```

```python
freq_df.sort_values('Roll').plot(x = 'Roll', kind='bar')
```

```python
#The first plot have the relation between rolls order and the result, and the second one the number of times that the result appears
```

# Challenge 2

```python
sum_rolls = 0
for i in dice_results['Roll']:
    sum_rolls += i
print(sum_rolls)
```

```python
sum_rolls
```

```python
freq_dict = {}
for i in dice_results['Roll']:
    if i not in freq_dict:
        freq_dict[i] = 1
    else:
        freq_dict[i] += 1
```

```python
print(freq_dict)
```

```python
freq_mean = 0
for i in list(zip(freq_dict,freq_dict.values())):
    freq_mean += i[0] * i[1]
freq_mean = freq_mean / len(dice_results['Roll'])
```

```python
freq_mean
```

```python
if not len(dice_results['Roll']) % 2:
    boring_median = dice_results['Roll'][len(sorted(dice_results['Roll']))/2] + dice_results['Roll'][(len(sorted(dice_results['Roll']))/2)+1] /2
else:
    boring_median = dice_results['Roll'][len(sorted(dice_results['Roll']))/2]
```
